```python
accern_data['date_time'] = pd.to_datetime(accern_data['date'], utc = True).dt.tz_convert('US/Eastern')
accern_data['day'] = accern_data['date_time'].dt.date
accern_data['hour'] = accern_data['date_time'].dt.hour
accern_data['week'] = accern_data['date_time'].dt.dayofweek
```

```python
accern_data = accern_data.sort_values(['ticker', 'date', 'hour'])
```

```python
full_data = pd.DataFrame(columns = ['ticker', 'day', 'hour'])
```

```python
ticker_list = []
for ticker in accern_data['ticker'].unique():
    ticker_list.extend([ticker] * 1718 * 24)
```

```python
day_list = []
for day in sorted(accern_data['day'].unique()):
    day_list.extend([day] * 24)
day_list = day_list * 19
```

```python
hour_list = sorted(accern_data['hour'].unique()) * 1718 * 19
```

```python
full_time = pd.DataFrame({'ticker' : ticker_list, 'day' : day_list, 'hour' : hour_list})
```

```python
full_time['dayofwork'] = pd.to_datetime(full_time['day']).dt.dayofweek + 1
```

```python
full_data = accern_data.drop('week', axis = 1).merge(full_time, how = 'right', on = ['ticker', 'day', 'hour'])
```

```python
full_data = full_data.sort_values(['ticker', 'day', 'hour'])
full_data = full_data.drop(['date', 'entity_share_class_figi', 'date_time'], axis = 1)
```

## Ffill

```python
full_data_ffill = full_data.fillna(method = 'ffill')
```

```python
full_data_ffill = full_data_ffill[full_data_ffill['day'] != datetime.date(2014,1,2)]
full_data_ffill.reset_index(inplace = True, drop = True)
```

```python
full_data_ffill = full_data_ffill[~full_data_ffill['day'].isin([datetime.date(2014,1,2), datetime.date(2018,9,13),\
                                                                datetime.date(2018,9,14), datetime.date(2018,9,15)])]
```

```python
full_data_ffill.to_csv('unagg_ffill_news.csv')
```

```python
#aggregation

full_data_ffill_copy = full_data_ffill.copy()
```

```python
full_data_ffill_copy.loc[full_data_ffill_copy['dayofwork'] == 6, 'day'] = \
full_data_ffill_copy.loc[full_data_ffill_copy['dayofwork'] == 6, 'day'] - pd.Timedelta(days = 1)

full_data_ffill_copy.loc[full_data_ffill_copy['dayofwork'] == 7, 'day'] = \
full_data_ffill_copy.loc[full_data_ffill_copy['dayofwork'] == 7, 'day'] - pd.Timedelta(days = 2)

full_data_ffill_copy.loc[full_data_ffill_copy['dayofwork'] == 6, 'hour'] = 8
full_data_ffill_copy.loc[full_data_ffill_copy['dayofwork'] == 7, 'hour'] = 8
```

```python
full_data_ffill_copy['day'] = np.where((full_data_ffill_copy['hour'] <= 7) & (full_data_ffill_copy['dayofwork'] != 1), \
                                       full_data_ffill_copy['day'] - pd.Timedelta(days = 1), \
                                       full_data_ffill_copy['day'])

full_data_ffill_copy['day'] = np.where((full_data_ffill_copy['hour'] <= 7) & (full_data_ffill_copy['dayofwork'] == 1), \
                                       full_data_ffill_copy['day'] - pd.Timedelta(days = 3), \
                                       full_data_ffill_copy['day'])
```

```python
full_data_ffill_copy.drop(['hour', 'dayofwork'], axis = 1, inplace = True)
```

```python
full_data_ffill_copy = full_data_ffill_copy.groupby(['ticker', 'day'], as_index = False).mean()
```

```python
full_data_ffill_copy = full_data_ffill_copy[full_data_ffill_copy['day'] != datetime.date(2014, 1, 2)]
```

```python
full_data_ffill_copy.to_csv('agg_ffill_news.csv')
```

```python
price_ticker_list = []
for ticker in price_data['ticker'].unique():
    price_ticker_list.extend([ticker] * 1226)
```

```python
price_date_list = sorted(full_data_ffill_copy['day'].unique()) *  19

price_full_time = pd.DataFrame({'ticker' : price_ticker_list, 'date' : price_date_list})
```

```python
price_full_time['day'] = price_full_time['date'].astype('str')
```

```python
price_full_data = price_data.merge(price_full_time, how = 'right', left_on = ['ticker', 'date'], right_on = ['ticker', 'day'])

price_full_data = price_full_data.sort_values(['ticker', 'day'])

price_full_data.reset_index(inplace = True, drop = True)
```

```python
price_data_ffill = price_full_data.fillna(method = 'ffill')
```

```python
price_data_ffill['open_price_change'] = (price_data_ffill['open'].shift(-2) - price_data_ffill['open'].shift(-1))\
                                        / price_data_ffill['open'].shift(-1)
```

```python
price_data_ffill.sort_values(['ticker', 'day'], inplace = True)
price_data_ffill = price_data_ffill[~price_data_ffill['day'].isin(['2014-01-02', '2018-09-13', '2018-09-14'])]
```
